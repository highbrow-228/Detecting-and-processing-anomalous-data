```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.metrics import classification_report

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE


df = pd.read_csv('..\\data\\creditcard.csv')
df
```

```python
# Розділяємо X, y
X = df.drop(['Class','Time','Amount'], axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

print(f"train: {X_train.shape}, test: {X_test.shape}")
```

```python
# SMOTE - StandardScaler - RandomForest
pipeline = Pipeline(
    [
        # Генерація синтетичних прикладів класу «fraud»
        ('smote', SMOTE(random_state=42)),

        # Нормалізація ознак
        ('scaler', StandardScaler()),

        # Класифікація випадковим лісом зі збалансованими підвибірками
        (
            'clf', RandomForestClassifier(
                n_estimators=100,
                class_weight='balanced_subsample',
                n_jobs=-1,
                random_state=42
            )
        )
    ]
)

# Навчання моделі
pipeline.fit(X_train, y_train)

# Оцінка якості класифікації на тесті
y_pred = pipeline.predict(X_test)
print(classification_report(y_test, y_pred, target_names=['normal','fraud']))
```

```python
# Несупервізований пошук аномалій
iso = IsolationForest(contamination=0.01, random_state=42)
iso.fit(X_train)

# Отримуємо anomaly score та прогноз
scores = iso.decision_function(X_test)
pred_if = iso.predict(X_test) # 1 — нормальне, -1 — аномалія

# Оцінка
y_if = (pred_if == -1).astype(int)
print(classification_report(y_test, y_if, target_names=['normal','fraud']))

# Візуалізуємо розподіл score
plt.figure()
plt.hist(scores, bins=50)
plt.title('Distribution of anomaly scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```
